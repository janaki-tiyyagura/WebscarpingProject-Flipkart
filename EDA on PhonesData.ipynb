```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('phonesdata.csv')
```

```python
data.head()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
fullname=data['Name'].to_list()
```

```python
nameandmodel=[]
colorandmemory=[]
```

```python
for each in fullname:
  nameandmodel.append(each.split('(')[0])
  colorandmemory.append(each.split('(')[1])
```

```python
name=[]
model=[]
color=[]
memory=[]
```

```python
for each in nameandmodel:
  name.append(each.split(" ",1)[0])
  model.append(each.split(" ",1)[1])
```

```python
for each in colorandmemory:
  color.append(each.split(",")[0])
  memory.append(each.split(",")[1].strip(")"))
```

```python
data1=pd.DataFrame({'Name':name,'Model':model,'Color':color,'Memory':memory})
```

```python
data1.head()
```

```python
data.drop('Name',axis=1,inplace=True)
```

```python
data.head()
```

```python
Memory=data['memory'].to_list()
```

```python
RAM=[]
ROM=[]
ExtendableMemory=[]
```

```python
for each in Memory:
  if len(each.split("|"))==3:
    RAM.append(each.split("|")[0])
    ROM.append(each.split("|")[1])
    ExtendableMemory.append(each.split("|")[2])
  else:
    RAM.append(each.split("|")[0])
    ROM.append(each.split("|")[1])
    ExtendableMemory.append(str(0))
```

```python
data2=pd.DataFrame({'RAM':RAM,'ROM':ROM,'ExtendableMemory':ExtendableMemory})
```

```python
data2.head()
```
